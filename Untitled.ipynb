```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: ₹"))

        if amount < 50:
            print("Invalid: Minimum recharge is ₹50. Try again.")
        elif amount in valid_plans:
            print("Recharge Successful!")
            break
        else:
            print("Invalid Plan! Available plans:", valid_plans)
```

```python
def inventory_alert(products):
    for item, stock in products.items():
        if stock < 15:
            print(f"{item}: Reorder Alert")
        else:
            print(f"{item}: Stock OK")

# Example
products = {
    "Pen": 10,
    "Notebook": 25,
    "Eraser": 5,
    "Marker": 18
}

inventory_alert(products)
```

```python
def student_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average} - Pass"
    else:
        return f"Average: {average} - Fail"

# Example
marks = [60, 45, 70, 50]
print(student_result(marks))
```

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)

    if peak.lower() == "yes":
        fare += fare * 0.25  # 25% extra

    return fare


while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    if distance <= 0:
        print("Invalid distance! Try again.")
    else:
        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare}")
        break
```

```python
def attendance_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage}% - Eligible"
    else:
        return f"Attendance: {percentage}% - Not Eligible"

# Example
attendance = ["P", "P", "A", "P", "P", "A", "P"]
print(attendance_eligibility(attendance))
```
