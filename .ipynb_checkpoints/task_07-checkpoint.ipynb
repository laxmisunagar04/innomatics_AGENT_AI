```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

```python
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# Sample Input
emp = EmployeeIDCard("Rahul", "EMP102", "IT")
emp.display_card()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# Sample Input
c1 = Contact("Anita", "9876543210")
c1.display_contact()
```

```python
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.product_name)
        print(f"Price: ₹{self.price}")


# Sample Input
p = Product("Headphones", 2499)
p.display_price_tag()
```

```python
class Movie:
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating

    def display_details(self):
        print("Movie:", self.movie_name)
        print(f"Rating: {self.rating} / 5")


# Sample Input
m = Movie("Inception", 4.8)
m.display_details()
```

```python
class DeliveryAddress:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


# Sample Input
d = DeliveryAddress("Suman", "Hyderabad")
d.display_delivery()
```
