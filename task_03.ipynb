```python
import string

query = "Buy mobile phone buy phone online"

# Step 1: Convert to lowercase
query = query.lower()

# Step 2: Remove punctuation
query = query.translate(str.maketrans('', '', string.punctuation))

# Step 3: Split into words and count frequency
word_count = {}
for word in query.split():
    word_count[word] = word_count.get(word, 0) + 1

# Step 4: Display only keywords searched more than once
result = {word: count for word, count in word_count.items() if count > 1}

print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# Step 1: Identify even readings and store (index, value) pairs
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

# Step 2: Display the result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Step 1: Count users per email domain
domain_count = {}
for email in emails:
    domain = email.split("@")[1]
    domain_count[domain] = domain_count.get(domain, 0) + 1

# Step 2: Calculate percentage usage of each domain
total_users = len(emails)
domain_percentage = {
    domain: (count / total_users) * 100
    for domain, count in domain_count.items()
}

# Step 3: Display the result
for domain, percentage in domain_percentage.items():
    print(f"{domain}: {int(percentage)}%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Step 1: Calculate the average daily sales
average_sales = sum(sales) / len(sales)

# Step 2: Define the spike threshold (30% above average)
threshold = average_sales * 1.30

# Step 3: Detect and display spike days
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Step 1: Count frequency of each user ID
id_count = {}
for uid in user_ids:
    id_count[uid] = id_count.get(uid, 0) + 1

# Step 2: Display only duplicate user IDs (count > 1)
for uid, count in id_count.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
