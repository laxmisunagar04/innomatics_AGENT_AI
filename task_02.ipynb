```python
numbers = [45, 22, 89, 10, 66]

max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
  
    if num < min_val:
        min_val = num
print("List:", numbers)
print("Max:", max_val)
print("Min:", min_val)
```

```python
text = "pythonp"
char_count = {}
for char in text:
    if char in char_count:
        char_count[char] += 1
    else:
        char_count[char] = 1
print(char_count)
```

```python
# Step 1: List of marks
marks = [70, 75, 80, 65]

# Step 2: Initialize total
total = 0

# Step 3: Calculate sum using loop
for m in marks:
    total += m

# Step 4: Count total number of marks
count = len(marks)

# Step 5: Calculate average
average = total / count

# Step 6: Display result
print("Average Marks:", average)
```
